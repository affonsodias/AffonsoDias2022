### Fazendo leitura de um arquivo CSV da base da dados da empresa Olist encontrada na plataforma Kaggle, uma empresa de e-commerce brasileira.

```python
# Importando a biblioteca Pandas do Python para manipulação de DataFrames
import pandas as pd
```

```python
# Extraindo os dados de um arquivo .csv com informações de registro das vendas 
pedidos = pd.read_csv("olist_orders_dataset.csv", sep = ',')
```

```python
pedidos
```

```python
# Renomeando as colunas

pedidos = pedidos.rename(columns = {'order_id': "ID do Pedido", 'customer_id': 'Cliente ID', 'order_status': 'Status do pedido', \
                          'order_purchase_timestamp': 'Data/hora da compra', 'order_approved_at': 'Data/hora do pedido aprovado'\
                         , 'order_delivered_carrier_date': 'Data/hora do pedido entregue na transportadora', \
                         'order_delivered_customer_date': 'Data/hora de entrega do pedido', 'order_estimated_delivery_date' \
                         : 'Data/hora prevista da entrega'})
```

```python
# Exibindo as 5 primeiras linhas do DataFrame

pedidos.head()
```

```python
# Modificando o status do pedido de "Delivered" para "Entregue"
pedidos['Status do pedido'] = pedidos['Status do pedido'].map({'delivered':'Entregue'})
```

```python
pedidos.head()
```

```python
# Lendo o arquivo .csv com os dados de registro das clientes 
clientes = pd.read_csv("olist_customers_dataset.csv", sep = ",")
```

```python
clientes
```

```python
# Renomeando as colunas

clientes = clientes.rename(columns = {'ID do Cliente': "Cliente ID", 'customer_unique_id':'ID exclusivo do Cliente',\
                                     'customer_zip_code_prefix':'Prefixo do código postal do Cliente', \
                                     'customer_city':'Cidade do Cliente','customer_state': 'Estado do Cliente'})
```

```python
clientes
```

```python
# Colocando as letras maiúsculas na coluna Cidade do Cliente
clientes['Cidade do Cliente'] = clientes['Cidade do Cliente'].str.upper()
```

```python
clientes
```

```python
# Criando um novo DataFrame com o relatório final das vendas, fazendo join dos DataFrames "Pedidos" e "Clientes"
resumo_venda = pedidos.merge(clientes, how='inner', on='Cliente ID')
```

```python
resumo_venda.head(10)
```

```python
#Excluindo a coluna "Id Exclusivo do Cliente"
resumo_venda = resumo_venda.drop(columns=['ID exclusivo do Cliente'])
```

```python
resumo_venda
```

```python
# Convertendo o tipo das colunas com data e hora para datetime
resumo_venda['Data/hora da compra'] = pd.to_datetime(resumo_venda['Data/hora da compra'],\
                                                     format="%d-%m-%Y %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda['Data/hora do pedido aprovado'] = pd.to_datetime(resumo_venda['Data/hora do pedido aprovado'],\
                                                              format="%Y-%m-%d %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda['Data/hora do pedido entregue na transportadora'] = pd.to_datetime(resumo_venda['Data/hora do pedido entregue na transportadora'],\
                                                                                format="%Y-%m-%d %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda['Data/hora de entrega do pedido'] = pd.to_datetime(resumo_venda['Data/hora de entrega do pedido'],\
                                                                format="%Y-%m-%d %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda['Prefixo do código postal do Cliente'] = pd.to_datetime(resumo_venda['Prefixo do código postal do Cliente'],\
                                                                     format="%Y-%m-%d %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda['Data/hora prevista da entrega'] = pd.to_datetime(resumo_venda['Data/hora prevista da entrega'],\
                                                               format="%Y-%m-%d %H:%M:%S").dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
resumo_venda
```

```python
# Fazendo a exportação do DataFrame para um arquivo do Excel
resumo_venda.to_excel('resumo_venda.xlsx', sheet_name = 'Resumo_vendas_Olist', index = 'False')
```
